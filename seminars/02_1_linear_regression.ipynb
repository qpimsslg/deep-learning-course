# 1. Линейная регрессия

Линейные методы предполагают, что между признаками объекта (features) и целевой переменной (target/label) существует линейная зависимость, то есть
$$y = w_1 x_1 + w_2 x_2 + ... + w_n x_n + b, $$
где
* $y$ – целевая переменная (то, что мы хотим предсказать),
* $x_1, \ldots, x_n$ – признаки объекта $\mathbf{x}$,
* $w_i$ – вес $i$-го признака, $b$ – bias (смещение, свободный член).

Часто предполагают, что объект $\mathbf{x}$ содержит в себе фиктивный признак равный 1 для представления свободного члена $b$. В этом случае формула принимает простой вид:
$$y = \langle \mathbf{w}, \mathbf{x} \rangle,$$
где $\langle \cdot, \cdot \rangle$ -- скалярное произведение векторов $\mathbf{w}$, $\mathbf{x} \in \mathbb{R}^n$. И тогда вектор $x = ( x_1, ..., x_n, 1)$

В многомерном случае формулу линейной регрессии можно переписать следующим образом:
$$ \mathbf{y} = X\mathbf{w}, $$
где
* $\mathbf{y}$ – столбец размера $\ell$ (количество объектов),
* $X$ – матрица признаков размера $\ell \times n$ (каждая строка матрицы является вектором признаков объекта),
* $\mathbf{w}$ – вектор весов размера $n$.

**Лосс:**
$$
\begin{aligned}
L(\mathbf{w}) &= \frac{1}{\ell}||\mathbf{y}_{pred} - \mathbf{y}||^2_2 =\\
&= \frac{1}{\ell}||X\mathbf{w} - \mathbf{y}||^2_2 = \frac{1}{\ell}\sum_{i=1}^{\ell}\left(\sum_{j=1}^{n} x_{ij}w_j - y_i\right)^2
\end{aligned}
$$

В данном случае функция потерь – это среднее квадрата отклонения. Норма в квадрате это как раз сумма квадратов компонент

По сути это выглядит следующим образом для многомерного случая:

### 1.1 Аналитическое решение

Минимизация ошибки по методу наименьших квадратов дает решение: $$ \mathbf{w} = (X^TX)^{-1}X^T \mathbf{y} $$

Реализуем класс линейной регрессии с помощью этой формулы. У нас изначально есть X, y

Мы должны внедрить наши веса w в нашу модель, поэтому это и просходит в ``` __init__```



Алгоритм которого мы придерживаемся:
1. Обработать X добавив вектор единиц (реализуется в методе fit)
2. Вычисление весов w по формуле $ \mathbf{w} = (X^TX)^{-1}X^T \mathbf{y} $ (реализуется в методе fit)
3. Формируем предсказание для тестовой выборки по формуле $ y = \langle \mathbf{w}, \mathbf{x} \rangle $ (реализуется в методе predict)

Метод ```get_weights``` используется для получения весов. Возвращаем копию весов, чтобы при их изменении, в самом классе веса не изменялись

Можно заметить, что мы нигде не подсчитываем функцию потерь $$ L(\mathbf{w}) = \frac{1}{\ell}\sum_{i=1}^{\ell}\left(\sum_{j=1}^{n} x_{ij}w_j - y_i\right)^2 $$

Это связано с тем, что тут мы используем аналитическое решение и подсчет лосса "вшит" в формулу выше

Но для ясности я решила привести вывод формулы ниже (тем более там не так много):


$$ L(\mathbf{w}) = \sum_{i=1}^{l} (\mathbf{x}_i^T - y_i) = (X\mathbf{w} - \mathbf{y})^T(X\mathbf{w} - \mathbf{y}) $$

$$ \frac{dL}{d\mathbf{w}} = 2X^T(X\mathbf{w} - \mathbf{y}) $$

Приравнивая градиент к нулю для нахождения минимума:
$$ 2X^T(X\mathbf{w} - \mathbf{y}) = 0 \implies X^T(X\mathbf{w} - \mathbf{y}) = 0 $$

Отсюда получаем аналитическое решение для весов $\mathbf{w}$:
$$ \mathbf{w} = (X^T X)^{-1} X^T \mathbf{y} $$

````python
class MyLinearRegression:
    def __init__(self):
        self.w = None

    def fit(self, X, y):
        '''Принимает на вход X, y и вычисляет веса по данной выборке'''
        # Не забудьте про фиктивный признак равный 1
        X = np.array(X) # l*n
        y = np.array(y) # l
        assert len(y.shape) == 1 and len(X.shape) == 2
        assert X.shape[0] == y.shape[0]

        # делаем размерность l, 1; по сути транспонировали вектор
        y = y[:, np.newaxis]
        l, n = X.shape

        # точно также в конец добавляем вектор из 1 длины l
        X_train = np.hstack((X, np.ones((l, 1))))

        # матричное умножение, применяем формулу выше
        self.w = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y

        return self

    def predict(self, X):
        '''Принимает на вход X и возвращает ответы модели'''
        # Не забудьте про фиктивный признак равный 1
        l, n = X.shape

        X_train = np.hstack((X, np.ones((l, 1))))

        y_pred = X_train @ self.w

        return y_pred

    def get_weights(self):
        return self.w.copy()
        # copy тут используется неспроста. Если copy не использовать, то get_weights()
        # выдаст ссылку на объект, а, значит, модифицируя результат применения функции
        # get_weights(), вы модифицируете и веса self.w. Если вы хотите модифицировать веса,
        # (например, в fit), используйте self.w
````

### 1.2. Градиентная оптимизация

Тут мы будем использовать другой способ предсказывать веса w. Для линейной регрессии решение задачи оптимизации можно выписать в явном виде. Но такое случается не для всех задач (это мы еще увидим позже). По этим причинам, мы воспользуемся методом градиентного спуска для оптимизации эмпирического риска.

Градиентый спуск заключается в следующих шагах:
1. Расчёт $\displaystyle\frac{\partial{L}}{\partial{\mathbf{w}}}$ – градиента функции потерь $L(\mathbf{w})$ – от значения параметров модели (весов $\mathbf{w}$);
В нашем случае это MSE
2. Шаг спуска – изменение весов $\mathbf{w}$ в сторону антиградиента с некоторым коэффициентом $\alpha$ (он же learning rate):
$$\mathbf{w} := \mathbf{w} - \alpha\frac{\partial{L}}{\partial{\mathbf{w}}};$$
3. Повторение п.1 и п.2 пока не наблюдается сходимость (изменения ошибки малы или отсутсвуют).



**Цель:** Реализуем новые классы линейной регрессии, в которой оптимизация проводится методами градиентного спуска.

**Предполагаемая зависимость:** $\mathbf{y} = X\mathbf{w}$, где $\mathbf{y} \in \mathbb{R}^{\ell\times 1}, X \in \mathbb{R}^{\ell\times n}, \mathbf{w} \in \mathbb{R}^{n\times 1}$

**Минимизируемая функция:**
$$
L(\mathbf{w}) = \frac{1}{\ell} ||\mathbf{y}_{pred} - \mathbf{y}||^2_2 = \frac{1}{\ell}||X\mathbf{w} - \mathbf{y}||^2_2
$$
Или в виде суммирования:
$$
L(\mathbf{w}) = \frac{1}{\ell}\sum_{i=1}^{\ell}\left(\sum_{j=1}^{n} x_{ij}w_j - y_i\right)^2
$$
А также в матричной форме:
$$
L(\mathbf{w}) = \frac{1}{\ell}(X\mathbf{w} - \mathbf{y})^T(X\mathbf{w} - \mathbf{y})
$$

**Градиент относительно $\mathbf{w}$:**
$$
\frac{\partial{L}}{\partial{\mathbf{w}}} = \frac{2}{\ell}(X^TX\mathbf{w} - X^T\mathbf{y}) = \frac{2}{\ell}X^T(\mathbf{y}_{pred} - \mathbf{y}) \,\, \in \mathbb{R}^{n \times 1}
$$

Мы создаем новый класс, который наследуется от класса линейной регрессии. Все что мы меняем – способ подсчета весов w, раньше мы это делали выразив w через матрицы, а теперь через градиентный спуск. Соответственно менять нужно только метод ```fit```

В самом ```fit``` мы подсчитываем градиент лосса, поэтому создадим новый метод ```_calc_gradient```, чтобы это можно было делать отдельно

````python
# класс наследуется от линейной регрессии
class MyGradientLinearRegression(MyLinearRegression):
    def __init__(self, **kwargs):
        # инициализация конструктора класса родителя
        super().__init__(**kwargs) # передает именные параметры родительскому конструктору

    def fit(self, X, y, lr=0.01, max_iter=100):
        ''' Принимает на вход X, y и вычисляет веса по данной выборке'''
        # Не забудьте про фиктивный признак равный 1
        X = np.array(X)
        y = np.array(y)
        assert len(y.shape) == 1 and len(X.shape) == 2
        assert X.shape[0] == y.shape[0]

        y = y[:, np.newaxis]
        l, n = X.shape

        # Добавляем признак из единиц
        X_train = np.hstack((X, np.ones((l, 1))))  # [l, n+1]

        # мы выбираем какие-то случайные веса. и будем спускаться отсюда
        self.w = np.random.randn(n + 1, 1)

        # функция потерь на разных итерациях
        self.losses = []

        for iter_num in range(max_iter):
            y_pred = self.predict(X)

            # Шаг 1
            # функция потерь в данный момент, 1 число
            # так как там суммируется по всем компонентам вектора
            self.losses.append(mean_squared_error(y, y_pred))

            grad = self._calc_gradient(X_train, y, y_pred)

            assert grad.shape == self.w.shape, \
            f"gradient shape {grad.shape} is not equal weight shape {self.w.shape}"

            # Шаг 2
            self.w -= lr * grad

        return self

    def _calc_gradient(self, X, y, y_pred):
        grad = (2. / X.shape[0]) * (X.T @ (y_pred - y))

        return grad

    def get_losses(self):
        return self.losses
````

#### Тестирование

Посчитаем предсказания на сгенерированном датасете

````python
regressor = MyGradientLinearRegression()

l = regressor.fit(X_train[:, np.newaxis], y_train, max_iter=100).get_losses()

predictions = regressor.predict(X_test[:, np.newaxis])
w = regressor.get_weights()
w
````

### 1.3 Стохастический градиентный спуск

Раньше функцию потерь мы считали на обучающей выборке на каждой итерации. Но это очень дорого делать на большой выборке (еще и каждую итерацию). Поэтому придумали такую методику

Стохастический градиентный спуск (SGD) – это улучшение обычного градиентного спуска, которое позволяет решить эту проблему.

На каждой итерации градиентного спуска будем считать градиент функции потерь не по всей выборке, а лишь по подмножеству выборки фиксированного размера $k$ (такое подмножество называется **батч**), и потом уже делать изменения по всей выборке используя такой "неточный" градиент. Существует несколько способов реализовать SGD:

Расчет идет тот же, но в X только те объекты, которые мы выбрали в батч $$
\frac{\partial{L}}{\partial{\mathbf{w}}}  = \frac{2}{\ell}X^T(\mathbf{y}_{pred} - \mathbf{y}) \,\, \in \mathbb{R}^{n \times 1}
$$
1. На каждом шаге спуска выбирать случайный батч.
image.png
2. Разделить обучение на эпохи, на каждой эпохе перемешивать выборку и проходиться батчами по всем элементам выборки.
Эпоха – один полный проход через всю обучающую выборку. Соответственно мы перемешиваем выборку случайным образом, делим выборку на батчи (по группам подряд) и проходим по ней итеративно до конца выборки

**Кратко** (благодарим Ольгу Липину, нашу слушательницу, которая написала этот комментарий к домашнему заданию):




> `yield` - это аналог `return`. Только `return` сразу выкидывает нас из функции, возвращая значение. А `yield` возвращает значение, и функция переходит в режим ожидания. И если мы вызовем её ещё раз, она продолжит работу с того же места, на котором остановилась.
>
> С помощью `yield` пишут генераторы. Конкретно здесь от нас хотят, чтоб мы написали генератор батчей. `yield` на каждом шаге возвращает `(batch_x, batch_y)`. И мы пишем цикл, который считает `batch_x`, `batch_y` и делает `yield batch_x, batch_y`. Таким образом, когда мы проходимся по объекту-генератору, функция на каждом шаге:
>
> 1. Возвращает нам батч.
> 2. Останавливается, запоминая предыдущее состояние.
> 3. На следующем шаге с текущего состояния, которое она помнит, возвращает новый батч

Если сейчас этот текст непонятен, ничего страшного. Вернитесь к нему, когда будете решать домашнее задание.


Мы будем пользоваться первым способом, потому что его проще реализовать. Добавим к нашему градиентному спуску сэмплирование батча, по которому будем считать градиент


Здесь мы заводим n_sample, здесь мы задаем размер батча
Далее мы подсчитываем градиент так, чтобы у нас выбирались случайные индексы, и количество выбранных элементов было равно n_sample

````python
class MySGDLinearRegression(MyGradientLinearRegression):
    def __init__(self, n_sample=10, **kwargs):
        super().__init__(**kwargs) # передает именные параметры родительскому конструктору
        self.w = None # это можно было и не писать по идее, так как self.w уже задан
        self.n_sample = n_sample

    def _calc_gradient(self, X, y, y_pred):
        # Главное отличие в SGD -- это использование подвыборки для шага оптимизации
        # выбираем индексы для нашей выборки, n_sample штук
        inds = np.random.choice(np.arange(X.shape[0]), size=self.n_sample, replace=False)

        grad = 2 / self.n_sample * X[inds].T @ (y_pred[inds] - y[inds])

        return grad
````

````python
plt.figure(figsize=(10, 6))

# смотрим разные размеры батчей
for ns in n_samples:
    l = MySGDLinearRegression(n_sample=ns).fit(
        X_train[:, np.newaxis],
        y_train,
        lr=5e-3,
        max_iter=150,
    ).get_losses()
    plt.plot(l, alpha=0.5, label=f'{ns} mini-batch size')

plt.title('Gradient descent learning')
plt.ylabel('loss')
plt.xlabel('iteration')

plt.legend()
plt.ylim((0, 150))
plt.grid(alpha=0.2)
plt.show()
````

Как видно по графикам, размер подвыборки влияет на стабильность сходимости (чем меньше, тем больше и резче изменения весов).
При этом количество итераций для минимизации примерно одинаково.

SGD также обычно улучшают адаптивным уменьшением величины шага (подробнее в курсах про методы оптимизаций). То есть чем больше итераций, тем меньше шаг для того, чтобы не перескочить минимум функции потерь (иначе при слишком больших шагах мы не сможем к нему приблизиться)
