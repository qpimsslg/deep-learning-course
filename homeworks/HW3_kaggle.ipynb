````python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
````

````python
#  1. ColumnTransformer       
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols), #  StandardScaler   
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols) #  OneHotEncoder   
    ])

#  2.   ,    
pipe_LR = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier', LogisticRegression(solver='liblinear'))])

#       LogisticRegression
#    'classifier',         
# param_grid = {
#     'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]
# }
param_grid = {
    'classifier__C': np.logspace(-4, 3, 50) #  np.logspace       
}

#   
data.dropna(inplace=True)

X_train = data.drop(columns=[target_col])
y_train = data[target_col]


#  3.  GridSearchCV     
CV_model = GridSearchCV(estimator=pipe_LR,
                            param_grid=param_grid,
                            cv=5,
                            scoring='roc_auc',
                            n_jobs=-1)

CV_model.fit(X_train, y_train)

#     
print("   Logistic Regression:", CV_model.best_params_)
print(" ROC-AUC  -  Logistic Regression:", CV_model.best_score_)

#  4.    
results_lr = CV_model.cv_results_
mean_scores_lr = results_lr['mean_test_score']
std_scores_lr = results_lr['std_test_score']
param_values_lr = results_lr['param_classifier__C'].data

#  4.  
plt.figure(figsize=(15, 8)) #   
plt.title(' ROC-AUC   C  Logistic Regression', fontsize=14)

plt.xscale('log')

plt.plot(param_values_lr, mean_scores_lr, label='  ROC-AUC', color='red', lw=3)

plt.fill_between(param_values_lr, mean_scores_lr - 2 * std_scores_lr, mean_scores_lr + 2 * std_scores_lr,
                 color='green', label=' ROC-AUC +- 2  ', alpha=0.5)

legend_box = plt.legend(framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")

plt.xlabel('  C ( )', fontsize=12)
plt.ylabel(' ROC-AUC  -', fontsize=12)

plt.ylim(min(mean_scores_lr - 2 * std_scores_lr) * 0.95, max(mean_scores_lr + 2 * std_scores_lr) * 1.05)

plt.grid(True)
plt.show()
````

````python
X_train, X_test, y_train, y_test = train_test_split(
                                                       data.drop(columns=[target_col]),
                                                       data[target_col],
                                                       train_size=0.8,
                                                       random_state=42
                                                       )

boosting_model = catboost.CatBoostClassifier(n_estimators=200,
                                             cat_features=cat_cols)

boosting_model.fit(X_train, y_train)

y_train_predicted = boosting_model.predict_proba(X_train)[:, 1]
y_test_predicted = boosting_model.predict_proba(X_test)[:, 1]

#  ROC-AUC    
roc_auc_train = roc_auc_score(y_train, y_train_predicted)
roc_auc_test = roc_auc_score(y_test, y_test_predicted)

print(f"ROC-AUC   : {roc_auc_train}")
print(f"ROC-AUC   : {roc_auc_test}")
````

````python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_test, y_test_predicted)

plt.figure(figsize=(10,7))

plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='validation AUC={:.4f}'.format(test_auc))

legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100), linestyle='--', color='gray', label='Random guess')
plt.title('ROC Curve')
plt.grid(True)
plt.show()
````
